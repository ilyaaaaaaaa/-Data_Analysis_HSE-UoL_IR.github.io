# A small manual on the main charts in seaborn

```python
import numpy as np # library for matrices and mathematics
import pandas as pd # library for working with tables

# libraries for visualization
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')    # chart style
%matplotlib inline 

import warnings
warnings.filterwarnings('ignore')
```

Next, we will use the built-in data set about tips.

```python
tips = pd.read_csv('tips.csv')
tips.head()
```

There are many different plotting functions in the Matplotlib and Seaborn libraries for data visualization. However, it is reasonable to use such functions not on any signs. We have already studied that signs (or variables) can be quantitative and categorical. It's very good if you know how to distinguish them. Depending on the types of features, there are their own charting functions. Let's look at the most frequently used ones on specific data.

#1. Signs one by one

## 1.1 Quantitative

Quantitative features can be displayed with the following useful functions:
* `histplot`
* `kdeplot`
* `rugplot`
* `boxplot`

### `histplot` , `kdeplot`

'histplot` with the parameter 'kde=True' shows a histogram and a graph of the distribution density

```python
sns.histplot(tips['total_bill'], kde = True);
```

We can only look at the density using the `kdeplot` function

```python
sns.kdeplot(tips['total_bill']);
```

### `rugplot`

'rugplot' shows the same as the distribution density graph, only in one-dimensional form. The denser the lines are, the higher the density.

```python
sns.rugplot(tips['total_bill']);
```

For clarity` 'rugplot' is often useful to use with other types of graphs.

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill']);
```

### `boxplot`

## 1.2 Categorical

### `countplot`

We count the number of values in each category.

```python
#sns.countplot(tips['sex']);
sns.countplot(x = 'sex', data = tips);
```

#2. Interaction of features

## 2.1 Quantitative with quantitative

### `kdeplot`

```python
sns.kdeplot(tips['total_bill'], tips['tip']);
```

### `jointplot`

"joint plot' shows a joint distribution across two variables. It has a `kind' parameter, which can take the following values:
* `scatter`
* `reg`
* `resid`
* `kde`
* `hex`

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'scatter');
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'reg');
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'resid');
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'kde');
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'hex');
```

## 2.2 Quantitative with categorical

### `barplot`

"barplot' aggregates data by the values of a categorical variable and applies a certain function to the values of the corresponding groups of a quantitative variable. By default, this is the **average** function, but this function can be changed in the `estimator` parameter

```python
sns.barplot(x = 'sex', y = 'total_bill', data = tips);
```

```python
sns.barplot(x = 'sex', y = 'total_bill', data = tips, estimator = np.var);
```

### `boxplot`

And again, the `boxplot' function we are already considering. With the interaction of features, this function can be many times more useful, since everything is clearer in comparison!

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips);
```

For convenience, you can flip

```python
sns.boxplot(x = 'total_bill', y = 'day', data = tips, orient = 'h');
```

Is it possible to open the third dimension in this way at all

```python
sns.boxplot(x = 'day', y = 'total_bill', hue = 'smoker', data = tips);
```

### `violinplot`

Performs the same function as 'boxplot'. In fact, these are two distribution density graphs rotated by 90 and -90 degrees, stuck together, resulting in something similar to a cello, hence the name.

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips);
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex');
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', split = True);
```

### `stripplot` , `swarmplot`

The 'strip plot' draws a scatter plot consisting of a single categorical variable.

```python
sns.stripplot(data = tips, x = 'day', y = 'total_bill');
```

```python
sns.stripplot(data = tips, x = 'day', y = 'total_bill', jitter = True, hue = 'sex');
```

```python
sns.stripplot(data = tips, x = 'day', y = 'total_bill', jitter = True, hue = 'sex', dodge = True);
```

'swarmplot' shows exactly the same thing, only the dots don't overlap

```python
sns.swarmplot(data = tips, x = 'day', y = 'total_bill', hue = 'sex', dodge = True);
```

Sometimes it is useful to combine these graphs with 'violinplot`

```python
sns.violinplot(data = tips, x = 'tip', y = 'day', palette = 'rainbow')
sns.swarmplot(data = tips, x = 'tip', y = 'day', size = 3);
```

## 2.3 Combinations of all features in the sample

### `pairplot`

'pairplot' shows the relationship between all pairs of variables. On the diagonal there are histograms, and in other places there are scatter diagrams.

```python
sns.pairplot(tips);
```

```python
sns.pairplot(tips, hue = 'sex');
```

### `PairGrid`

"Airgrid` is a more advanced function that allows you to show the relationship between all pairs of parameters for various functions. In fact, `pairplot' is a special case of it.

```python
sns.PairGrid(tips);
```

```python
g = sns.PairGrid(tips)
g.map(plt.scatter);
```

```python
g = sns.PairGrid(tips)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot);
```

##2.4 Quantitative and somewhat categorical

### `Facet Grid`
'FacetGrid' allows you to visualize the joint distribution of individual features of several variables

```python
g = sns.FacetGrid(tips, col = 'time', row = 'smoker');
```

```python
g = sns.FacetGrid(tips, col = 'time', row = 'smoker');
g = g.map(plt.hist, 'total_bill');
```

```python
g = sns.FacetGrid(tips, col = 'time', row = 'smoker', hue = 'sex');
g = g.map(plt.scatter, 'total_bill', 'tip').add_legend();
```
